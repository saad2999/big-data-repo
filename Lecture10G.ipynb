```python
'''
How to handle missing values.....
    1. Replace with means(average) value..... 
    2. Replace with Maximum/minimum value......
    3. Replace with maximum frequent (most occurred) value....
    4. Delete Record
'''
```

```python
df['normalized-losses'].head(25)
```

```python
#Average -- Mean.....
avg = df['normalized-losses'].astype('float').mean(axis=0)
```

```python
avg
```

```python
df['normalized-losses'].replace(np.nan, avg, inplace=True)
```

```python
df['normalized-losses'].isnull().sum()
```

```python
m = df['num-of-doors'].value_counts().idxmax()
```
