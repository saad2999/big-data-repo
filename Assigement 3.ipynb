```python
import pandas as pd
import random
x=pd.read_csv(r'C:\Users\nopsh\Downloads\marks.csv')
```

```python
print('How many ilteration do u need ')
a=input()
for d in range(int(a)):
    if d==0:
        n1=random.choices(x['marks'])
        n2=random.choices(x['marks'])
        x['Ilteration '+str(d+1)+' C1']=abs(x['marks']-n1)
        x['Ilteration '+str(d+1)+' C2']=abs(x['marks']-n2)
        C1=0
        C2=0
        c1No=0
        c2No=0
        for xl in range(8):
            if x['Ilteration '+str(d+1)+' C1'][xl]<x['Ilteration '+str(d+1)+' C2'][xl]:
                C1=C1+x['marks'][xl]
                c1No=c1No+1
            else:
                C2=C2+x['marks'][xl]
                c2No=c2No+1
        n1=C1/c1No
        n2=C2/c2No
    else:
        x['Ilteration '+str(d+1)+' C1']=abs(x['marks']-n1)
        x['Ilteration '+str(d+1)+' C2']=abs(x['marks']-n2)
        C1=0
        C2=0
        c1No=0
        c2No=0
        for xl in range(8):
            if x['Ilteration '+str(d+1)+' C1'][xl]<x['Ilteration '+str(d+1)+' C2'][xl]:
                C1=C1+x['marks'][xl]
                c1No=c1No+1
            else:
                C2=C2+x['marks'][xl]
                c2No=c2No+1
        n1=C1/c1No
        n2=C2/c2No
```
